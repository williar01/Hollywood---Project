```python
import panel as pn
pn.extension('plotly')
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
import numpy as np
import os
from pathlib import Path
from dotenv import load_dotenv
import warnings
warnings.filterwarnings('ignore')
import plotly.express as px
```

```python
file_path_imdb = Path("Data/imdb_modified.csv")
imdb_modified_data = pd.read_csv(file_path_imdb)
imdb_modified_data.head()
```

```python
imdb_modified_data_copy =imdb_modified_data.drop_duplicates()
```

```python
imdb_modified_data_copy=imdb_modified_data_copy[imdb_modified_data_copy.index.notnull()].reset_index()
imdb_faces =imdb_modified_data_copy[['title_year', 'facenumber_in_poster', 'movie_title','imdb_score']]
imdb_faces.head()
```

```python
#I located the movies that weren't merging so I can rename them so they can be analyzed.

imdb_faces.at[237,'movie_title'] = 'Star Wars Ep. I: The Phantom Menace'
imdb_faces.at[14,'movie_title'] = 'Pirates of the Caribbean: Dead Man’s Chest'
imdb_faces.at[233,'movie_title'] = 'Star Wars Ep. III: Revenge of the Sith'
imdb_faces.at[200,'movie_title'] = 'Harry Potter and the Sorcerer’s Stone'
```

```python
file_path_top = Path("Data/HighestGrossers.csv")
highest_grossers_copy = pd.read_csv(file_path_top, index_col="YEAR")
highest_grossers_copy.head()
```

```python
#merges the 2 data sets
imdb_faces['MOVIE'] = imdb_faces["movie_title"].str.rstrip()

movie_data_faces=pd.merge(
imdb_faces,
highest_grossers_copy,
on='MOVIE'
)
```

```python
#cleaning and arranging the data 
movie_data_faces = movie_data_faces.drop_duplicates()
movie_data_faces.at[4,'title_year']= 1998
movie_data_faces.at[2,'title_year']= 2006
movie_data_faces['title_year'] = movie_data_faces['title_year'].astype(int)
```

```python
#highest grossing flim per year

movie_data_faces = movie_data_faces.drop(columns = 'movie_title')
movie_data_faces = movie_data_faces.set_index('title_year')
movie_data_faces = movie_data_faces.sort_index()
```

```python
#fixing some of the data, to get an accruate analysis

#replaces specific values that I wanted to check 
movie_data_faces.at[1998,'facenumber_in_poster']= 2.0
movie_data_faces.at[1997,'facenumber_in_poster']= 2.0
movie_data_faces.at[2000,'facenumber_in_poster']= 1.0
movie_data_faces.at[2012,'facenumber_in_poster']= 6.0
movie_data_faces.at[2013,'facenumber_in_poster']= 5.0
movie_data_faces.at[1999,'facenumber_in_poster']= 4.0
movie_data_faces.at[2001,'facenumber_in_poster']= 5.0
movie_data_faces.at[2006,'facenumber_in_poster']= 3.0
```
